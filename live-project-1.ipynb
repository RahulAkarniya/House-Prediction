# DROP LATITUDE OR LONGITUTE

```python
train.drop(['LONGITUDE','LATITUDE'], axis = 1, inplace = True)
print(train.shape)
```

```python
test.drop(['LONGITUDE','LATITUDE'], axis = 1, inplace = True)
print(test.shape)
```

```python
train.head()
```

```python
train = pd.get_dummies(train, columns=["POSTED_BY", 'BHK_OR_RK'], prefix=["POSTED_BY", 'IS_'] )
```

```python
test = pd.get_dummies(test, columns=["POSTED_BY", 'BHK_OR_RK'], prefix=["POSTED_BY", 'IS_'] )
```

```python
train.head()
```

```python
train.drop(['ADDRESS'], axis = 1, inplace = True)
test.drop(['ADDRESS'], axis = 1, inplace = True)
```

```python
train.head()
```

```python
X= train.drop(columns = ['TARGET(PRICE_IN_LACS)'], axis=1)
y= train['TARGET(PRICE_IN_LACS)']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test= train_test_split(X,y, test_size= 0.75, random_state= 0)
```

```python
X_train,X_test,y_train,y_test= train_test_split(X,y,test_size=0.2,random_state=2)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score

rf_model = RandomForestRegressor(n_estimators=200, random_state=42)

rf_model.fit(X_train, y_train)

preds = rf_model.predict(X_test)

print('Random Forest: ', r2_score(y_test, preds))
```

```python
from sklearn.linear_model import LinearRegression
model= LinearRegression()
model.fit(X_train, y_train)
```

```python
y_pred= model.predict(X_test)
print(y_pred)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test, y_pred)* 100
```

```python
accuracy= model.score(X_test, y_test)
print(accuracy*100, "%")
```
